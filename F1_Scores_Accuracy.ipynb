```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
scores_ideology = {
    "Model": ["Linear SVC Text-Only", "Variable Selected SVC", "Gradient Boosting", "Support Vector Classifier", "Logistic Regression", "Random Forest", "Decision Tree", "Naive Bayes", "Neural Network"],
    "Accuracy": [0.45, 0.54, 0.45, 0.45, 0.48, 0.43, 0.45, 0.48, 0.43],
    "F1 Score (Weighted)": [0.43, 0.54, 0.35, 0.44, 0.46, 0.43, 0.45, 0.43, 0.39]
}

scores_party = {
    "Model": ["Linear SVC Text-Only", "Variable Selected SVC", "Gradient Boosting", "Support Vector Classifier", "Logistic Regression", "Random Forest", "Decision Tree", "Naive Bayes", "Neural Network"],
    "Accuracy": [0.16, 0.62, 0.24, 0.23, 0.25, 0.22, 0.20, 0.26, 0.20],
    "F1 Score (Weighted)": [0.15, 0.34, 0.20, 0.19, 0.21, 0.20, 0.19, 0.12, 0.17]
}

df_i = pd.DataFrame(scores_ideology)
df_p = pd.DataFrame(scores_party)

```

```python
df_i
df_p
```

# Party

```python
# Create the figure and axis
plt.figure(figsize=(10, 6))

# Create a color palette (customize these colors as needed)
colors = ['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728', 
          '#9467bd', '#8c564b', '#e377c2', '#7f7f7f',
          '#bcbd22', '#17becf', '#aec7e8', '#ffbb78']

# Create the bar plot
bars = plt.bar(df_p['Model'], df_p['Accuracy'], color=colors)

# Add a dashed line at 0.5
plt.axhline(y=0.5, color='gray', linestyle='--', linewidth=1)

# Add value labels on top of each bar
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2., height,
             f'{height:.2f}',
             ha='center', va='bottom')

# Customize the plot
plt.title('Party Choice Model Comparison', fontsize=14)
plt.ylabel('Accuracy')
plt.ylim(0, 0.7)  # Adjust based on your data range
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better readability

# Show the plot
plt.tight_layout()
plt.show()
```
